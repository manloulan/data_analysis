```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression

titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')

titanic_train['Sex'] = titanic_train['Sex'].replace({"male":1,'female':0})
titanic_train['Embarked'] =titanic_train['Embarked'].replace({'S':0 ,'Q':1,'C':2})
titanic_test['Sex'] = titanic_test['Sex'].replace({"male":1,'female':0})
titanic_test['Embarked'] = titanic_test['Embarked'].replace({'S':0 ,'Q':1,'C':2})
titanic_train['Age'] = titanic_train['Age'].fillna(titanic_train['Age'].mean())
titanic_train['Embarked'] =titanic_train['Embarked'].fillna(titanic_train['Embarked'].mean())
titanic_test['Embarked'] = titanic_test['Embarked'].fillna(titanic_train['Embarked'].mean())
titanic_test['Age'] = titanic_test['Age'].fillna(titanic_train['Age'].mean())
# print(titanic_test)
x = pd.get_dummies(titanic_train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']])
y = titanic_train["Survived"]
X_test_final = pd.get_dummies(titanic_test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']])
X_test_final = X_test_final.fillna(X_test_final.mean())
model = LogisticRegression(max_iter=200)
model.fit(x, y)

predictions = model.predict(X_test_final)

output = pd.DataFrame({'PassengerId': titanic_test['PassengerId'], 'Survived': predictions})
print(output)
output.to_csv('submission2.csv', index=False)
print("Your submission was successfully saved!")
```
